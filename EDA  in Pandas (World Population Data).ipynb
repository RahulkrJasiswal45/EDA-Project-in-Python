```python
df['2023 population'] = df['2023 population'].astype(np.float64)
df['2022 population'] = df['2022 population'].astype(np.float64)
df['2020 population'] = df['2020 population'].astype(np.float64)
df['2015 population'] = df['2015 population'].astype(np.float64)
df['2010 population'] = df['2010 population'].astype(np.float64)
df['2000 population'] = df['2000 population'].astype(np.float64)
df['1990 population'] = df['1990 population'].astype(np.float64)
df['1980 population'] = df['1980 population'].astype(np.float64)
df['1970 population'] = df['1970 population'].astype(np.float64)
df['density (km²)'] = df['density (km²)'].astype(np.float64)
df
```

```python
#df['cca3'] = df['cca3'].str.replace('%', '').astype(float)
df['growth rate'] = df['growth rate'].str.replace('%', '').astype(float)
df['world percentage'] = df['world percentage'].str.replace('%', '').astype(float)
df.dtypes
```
